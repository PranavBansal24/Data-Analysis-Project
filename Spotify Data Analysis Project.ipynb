```python
corr_map=data.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_map, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap='mako')
heatmap.set_title("Correlation Heatmap Between Variables ")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_data = data.sample(int(0.005*len(data)))
```

```python
print(len(sample_data))
```

```python
plt.figure(figsize=(20,6))
sns.regplot(data=sample_data, x="energy", y="loudness" , color="red").set(title="Energy vs Loudness Correlation")
# positive correlation plotting
```

```python
plt.figure(figsize=(15,6))
sns.regplot(data=sample_data , x="acousticness" , y="popularity", color="orange").set(title="Acousticness vs Popularity Correlaton")
# Negative correlation plotting
```

```python
data['years']=data.index.get_level_values('release_date')
data.years=pd.to_datetime(data.years)
years=data.years.dt.year
```

```python
sns.displot(years,discrete=True, aspect=2, height=7, kind='hist', color='cyan').set(title="No of songs released per year")
```

```python
total_dur=data.duration_m
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y=total_dur, ax=ax, errwidth=False).set(title="Years vs Duration BarPlot")
plt.xticks(rotation=90)
```

```python
total_dur=data.duration_m
sns.set_style(style="whitegrid")
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x=years, y=total_dur, ax=ax).set(title="Years vs Duration LinePlot")
plt.xticks(rotation=60)
```
